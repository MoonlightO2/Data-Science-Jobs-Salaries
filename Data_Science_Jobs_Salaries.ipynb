# **Data Science Jobs Salaries**
Dataset: https://www.kaggle.com/datasets/saurabhshahane/data-science-jobs-salaries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file = '/content/drive/MyDrive/Projects/Portfolio/Jobs/Data Science jobs/Data Science Jobs Salaries.csv'
```

```python
data = pd.read_csv(file)
```

```python
data.rename(columns=lambda x: x.replace('_', ' ').title(), inplace=True)
print('Shape: ', data.shape, '\n')
print('Columns: ', data.columns.tolist(), '\n')
data.head()
```

```python
for column_name, values in data.items():
    unique_values = set(values)
    num_unique_values = len(unique_values)

    print(f"Column: {column_name}")
    print(f"Unique Values: {unique_values}")
    print(f"Number of Unique Values: {num_unique_values}\n")
```

```python
# Sort DataFrame by salary in USD in descending order
sorted = data.sort_values(by='Salary In Usd', ascending=False)

# Get top 20 job titles
top_20 = sorted.head(50)

# Plotting
plt.figure(figsize=(10, 10))
plt.barh(top_20['Job Title'], top_20['Salary In Usd'], color='skyblue')
plt.xlabel('Salary in USD')
plt.ylabel('Job Title')
plt.title('Top 20 Job Titles by Salary (USD)')
plt.gca().invert_yaxis()  # Invert y-axis to have highest salary at the top
plt.show()
```

### 1. **Work Year**: We can visualize the distribution of work years using a histogram.

```python
plt.hist(data['Work Year'], bins=20, color='skyblue')
plt.xlabel('Work Year')
plt.ylabel('Frequency')
plt.title('Distribution of Work Year')
plt.show()
```

### 2. **Experience Level**: We can create a bar plot to show the count of each experience level.

```python
sns.countplot(data['Experience Level'])
plt.xlabel('Experience Level')
plt.ylabel('Count')
plt.title('Count of Experience Levels')
plt.xticks(rotation=45)
plt.show()
```

### 3. **Employment Type**: We can visualize the distribution of employment types using a pie chart.

```python
plt.pie(data['Employment Type'].value_counts(), labels=data['Employment Type'].value_counts().index, autopct='%1.1f%%')
plt.title('Distribution of Employment Types')
plt.show()
```

### 4. **Job Title**: We can create a word cloud to visualize the most common job titles.

```python
from wordcloud import WordCloud

wordcloud = WordCloud(width=400, height=400, background_color='black').generate(' '.join(data['Job Title']))
plt.figure(figsize=(5, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Word Cloud of Job Titles')
plt.show()
```

### 5. **Salary**: We can visualize the distribution of salaries using a histogram.

```python
plt.hist(data['Salary'], bins=20, color='lightgreen')
plt.xlabel('Salary')
plt.ylabel('Frequency')
plt.title('Distribution of Salary')
plt.show()
```

### 6. **Salary Currency**: We can create a bar plot to show the count of each currency.

```python
sns.countplot(data['Salary Currency'])
plt.xlabel('Salary Currency')
plt.ylabel('Count')
plt.title('Count of Salary Currencies')
plt.xticks(rotation=45)
plt.show()
```

### 7. **Salary In USD**: We can visualize the distribution of salaries converted to USD using a histogram.

```python
plt.hist(data['Salary In Usd'], bins=20, color='lightcoral')
plt.xlabel('Salary (USD)')
plt.ylabel('Frequency')
plt.title('Distribution of Salary in USD')
plt.show()
```

### 8. **Employee Residence**: We can create a bar plot to show the count of employees residing in each location.

```python
plt.figure(figsize=(10, 6))
sns.countplot(y=data['Employee Residence'], order=data['Employee Residence'].value_counts().index, palette='viridis')
plt.xlabel('Count')
plt.ylabel('Employee Residence')
plt.title('Count of Employee Residence')
plt.show()
```

### 9. **Remote Ratio**: We can visualize the distribution of remote ratios using a histogram.

```python
plt.hist(data['Remote Ratio'], bins=20, color='orange')
plt.xlabel('Remote Ratio')
plt.ylabel('Frequency')
plt.title('Distribution of Remote Ratio')
plt.show()
```

### 10. **Company Location**: We can create a bar plot to show the count of companies in each location.

```python
plt.figure(figsize=(10, 6))
sns.countplot(y=data['Company Location'], order=data['Company Location'].value_counts().index, palette='plasma')
plt.xlabel('Count')
plt.ylabel('Company Location')
plt.title('Count of Company Location')
plt.show()
```

### 11. **Company Size**: We can create a bar plot to show the count of companies in each size category.

```python
sns.countplot(data['Company Size'])
plt.xlabel('Company Size')
plt.ylabel('Count')
plt.title('Count of Company Size')
plt.xticks(rotation=45)
plt.show()
```
